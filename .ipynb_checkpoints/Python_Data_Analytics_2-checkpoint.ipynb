### Import Libraries

### Line graph

```python
# Import data source
population = pd.read_csv('POP.csv')

# Check to see if there are any null values
# population.info()

population['date'] = pd.to_datetime(population['date'])
population['value'] = population['value'] / 1000

plt.figure(figsize=(8,8))
plt.title('Population of the United States (1952-2019)', fontdict={'fontweight':'bold'})

plt.plot(population.date.dt.year, population.value, color='#6789bf')

plt.xlabel('Year')
plt.ylabel('Population (in millions)')

population['date'] = population['date'].astype(int)


plt.show()

```

### Histogram 

```python
ramen = pd.read_csv('ramen-ratings.csv')

#Removing unrated data
ramen.drop(ramen.index[ramen['Stars'] == 'Unrated'], inplace = True)

#convert all star ratings to float first (because data is strings) 
#Then convert to int (to round to nearest whole number). 

ramen['Stars'] = ramen['Stars'].astype(float)
ramen['Stars'] = ramen['Stars'].astype(int) 

bins = [0,1,2,3,4,5,6]

plt.hist(ramen.Stars, bins=bins, color='#8796ad')

plt.ylabel('Number of Ratings')
plt.xlabel('Stars')
plt.title('Ramen Ratings')
plt.xticks(bins)

plt.show()
```

### Pie Chart

```python
#ramen.Style.unique()

Cup = ramen.loc[ramen['Style'] == 'Cup'].count()[0]
Pack = ramen.loc[ramen['Style'] == 'Pack'].count()[0]
Tray = ramen.loc[ramen['Style'] == 'Tray'].count()[0]
Bowl = ramen.loc[ramen['Style'] == 'Bowl'].count()[0]
Box = ramen.loc[ramen['Style'] == 'Box'].count()[0]
Can = ramen.loc[ramen['Style'] == 'Can'].count()[0]
Bar = ramen.loc[ramen['Style'] == 'Bar'].count()[0]

Other = Box + Can + Bar

plt.figure(figsize=(8,5), dpi=100)

labels = ['Cup', 'Pack', 'Tray', 'Bowl', 'Other']
colors = ['#dbfffe', '#a2e8e6', '#7fd4d1', '#54a8a5', '#FFFFFF']


plt.pie([Cup, Pack, Tray, Bowl, Other],
       labels= labels, colors = colors, autopct='%.2f %%', pctdistance = 0.75)

plt.title('Types of Ramen')

plt.show()
```

### Bar Chart

```python
wine = pd.read_csv('winemag-data-130k-v2.csv')

plt.figure(figsize=(10,5))

country_dict = dict.fromkeys(wine.country.unique() , 1)


for c in country_dict:
    country_dict[c] = (np.where(wine['country']==c, wine['variety']=='Chardonnay',0).sum())

country_dict = {key:val for key, val in country_dict.items() if val > 100} 


fig = sns.barplot(list(country_dict.keys()), list(country_dict.values()))
plt.xlabel("Country")
plt.ylabel("Count of Chardonnay reviews")
plt.title('Countries with over 100 Chardonnay reviews')
plt.show(fig)

fig.figure.savefig('wine.png', dpi=100)

```
